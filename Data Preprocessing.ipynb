# DATA PREPROCESSING

# MinMaxScaler

Rumus minmax scaler :
    x = (x(old) - x(min)) / (x(max) - x(min))
    
    x(old) = Merupakan nilai awal / lama
    x(min) = Merupakan nilai minimum atau terkecil
    x(max) = Merupakan nilai maximum atau terbesar

*Mengambil kolom dataframe

```python
cols = dftrain.columns[2:]
cols
```

*Menyalin dataframe dan diolah menjadi sekala MinMax

```python
dfmms = dftrain.copy()
for a in cols:
    dfmms[a] = (dftrain[a] - dftrain[a].min()) / (dftrain[a].max() - dftrain[a].min())
```

```python
dfmms.tail()
```

#  Z score

Rumus Z score :
    x = (x(old) - x(mean)) / x(std)
    
    x(old) = Merupakan nilai awal / lama
    x(mean) = Merupakan rata-rata nilai
    x(std) = Merupakan standart deviasi nilai

*Mengambil nilai std yang disediakan oleh pandas

```python
dftrain.describe()
```

*Atau mencari sendiri Standart deviasi

```python
np.sqrt(np.sum((dftrain['Z1'] - dftrain['Z1'].mean())**2) / (len(dftrain['Z1']) - 1))
```

Rumus Standard Deviasi :

x = akar((x - rata2)^2 / jumlah)

.

*Eksekusi rumus ZScore

```python
dfz = dftrain.copy()
for i in cols:
    df_std = np.sqrt(np.sum((dftrain[i] - dftrain[i].mean())**2) / (len(dftrain[i]) - 1))
    
    
    dfz[i] = (dftrain[i]-dftrain[i].mean()) / df_std
```
